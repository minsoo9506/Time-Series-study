## 2.4 Seasonal plots

```python
# using pandas -> change 'object' to 'datetime'

data['Date'] = pd.to_datetime(data['Date'])
data['Date'].head(3)
```

```python
# pd.Timestamp -> year, month, day
data['year'] = data['Date'].map(lambda x : x.year)
data['month'] = data['Date'].map(lambda x : x.month)
data['day'] = data['Date'].map(lambda x : x.day)
```

```python
for i in range(3):
    data.groupby(['year','month'])['Close'].sum()[0+12*i:11+12*i].plot()
```
